```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sin(x)', color='blue')
plt.plot(x, y2, label='Cos(x)', color='red', linestyle='-.')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

```python
# Parametric equations for the heart shape
t = np.linspace(0, 2*np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Plot the heart shape
plt.plot(x, y, color='r')
plt.axis('equal')
plt.title('Heart Shape')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D']
x = np.arange(len(categories))
heights = [3, 7, 5, 2]
widths = [0.5, 0.6, 0.4, 0.7]
bottoms = [1, 2, 0, 3]
colors = ['red', 'green', 'blue', 'orange']
edge_colors = ['black', 'black', 'black', 'black']
line_widths = [1, 2, 1.5, 2.5]
labels = ['Group A', 'Group B', 'Group C', 'Group D']

# Create bar chart
plt.bar(x, heights,
        width=widths,
        bottom=bottoms,
        align='center',
        color=colors,
        edgecolor=edge_colors,
        linewidth=line_widths,
        tick_label=categories,
        label=labels)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart')
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50, 300, size=50)        # s: marker sizes
colors = np.random.rand(50)                        # c: color values
edge_colors = ['black' if i % 2 == 0 else 'white' for i in range(50)]

# Create scatter plot
scatter = plt.scatter(
    x, y,
    s=sizes,                         # marker size
    c=colors,                        # color values
    marker='o',                      # marker type (circle)
    cmap='viridis',                  # colormap for color values
    norm=None,                       # color normalization (default)
    vmin=0, vmax=1,                  # color scale range
    alpha=0.7,                       # transparency
    linewidths=1.5,                  # edge width
    edgecolors=edge_colors,         # edge color
    label='Data Points'             # for legend
)

# Additional plot settings
plt.colorbar(scatter, label='Color scale')  # show color bar
plt.title("Full Parameter Usage in plt.scatter()")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.axis('equal')

plt.show()
```

```python
plt.pie(x=[2,1,3])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
sizes = [25, 35, 20, 20]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # explode first slice

# Create pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    labeldistance=1.1,
    startangle=90,
    radius=1,
    counterclock=False,
    wedgeprops={'linewidth': 1, 'edgecolor': 'black'},
    textprops={'color': 'black'},
    center=(0, 0),
    frame=True,
    rotatelabels=False
)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add legend
ax.legend(wedges, labels, title="Fruits", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Display the plot
plt.show()
```
