## Let's work with matplotlib

```python
from matplotlib import pyplot as plt 
```

you can make an empty plot using "plt.plot()"

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3],[2,4,6])
plt.show()
```

tipðŸ˜Ž: "plot" method has line style as default

### title in matplotlib

```python
plt.plot([1,2,3],[2,4,6])

plt.title("line plot")      # main title
plt.show()
```

```python
plt.plot([1,2,3],[2,4,6])

plt.title("line plot")      # main title
plt.xlabel("x axis")        # x axis title
plt.ylabel("y axis")        # y asix title
plt.show()
```

### changing color

```python
plt.plot([1,2,3,4,5,6,7,8], [1,5,4,5,2,1,6,3], color='g')
plt.show()
```

### more than one line!

```python
plt.plot([1,2,3,4,5,6,7,8], [1,5,4,5,2,1,6,3], label="line 1")
plt.plot([1,2,3,4,5,6,7,8], [4,3,1,6,2,5,3,7], label="line 2")

plt.title("line plot with two lines")    
plt.xlabel("x axis")       
plt.ylabel("y axis")

plt.legend()                  #showing label of lines
plt.show()
```

tipðŸ˜‰: You can add as many lines as you want to the line plot using plt.plot()

### numpy and matplotlib

numpy and matplotlib have great cooperate together and you can do a lot of things with them

```python
import numpy as np 

# Create an ndarray on x axis using the numpy arange() function:
x = np.arange(-10,11)
# Store equation values on y axis:
# y = 2*x + 10
y = x**2 + 5

plt.title("NumPy Array Plot")
plt.plot(x,y)
plt.show()
```

You can visualize anything you have in mindâœ¨ðŸ˜

```python
# Parametric equations for the heart shape
t = np.linspace(0, 2*np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Plot the heart shape
plt.plot(x, y, color='r')
plt.axis('equal')
plt.title('Heart Shape')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
